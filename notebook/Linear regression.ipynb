```python
df['total_sqft'].to_list()
```

```python
def process_total_sqft(r):
    if '-' in r:
        value = (float(r.split('-')[0]) + float(r.split('-')[1]))/2
        return value
    else:
        return r
```

```python
df['total_sqft'] = df['total_sqft'].map(process_total_sqft)
```

```python
df['total_sqft'].to_list()
```

```python
import re 
t = re.findall(r'\d+', "2323kjhndv s.meret")
```

```python
def test(r):
    r = str(r)
    t = re.findall(r'\d+',r)
    return t[0]
```

```python
df['total_sqft']=df['total_sqft'].map(test)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['price'],axis=1)
```

```python
y = df['price']
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
model = lr.fit(X_train,y_train)
```

```python
model
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```
